```python
class DataAnalyzer:
    def __init__(self):
        """Initializes an empty set and an empty dictionary."""
        self.data_set = set()
        self.data_dict = {}

    def add_to_set(self, elements):
        """Adds a list of elements to the set."""
        self.data_set.update(elements)

    def remove_from_set(self, element):
        """Removes an element from the set if it exists."""
        self.data_set.discard(element)

    def get_set(self):
        """Returns the current state of the set."""
        return self.data_set

    def create_dictionary(self, keys, values):
        """Creates a dictionary from two lists of equal length, one for keys and one for values."""
        if len(keys) == len(values):
            self.data_dict = dict(zip(keys, values))
        else:
            print("Error: Keys and values lists must have the same length.")

    def update_dictionary(self, key, value):
        """Adds a key-value pair to the dictionary."""
        self.data_dict[key] = value

    def get_dictionary(self):
        """Returns the current state of the dictionary."""
        return self.data_dict

    def search_dictionary(self, key):
        """Checks if a key exists in the dictionary."""
        return key in self.data_dict

    def remove_from_dictionary(self, key):
        """Removes a key-value pair from the dictionary if the key exists."""
        if key in self.data_dict:
            del self.data_dict[key]
        else:
            print("Key not found.")
```

```python
analyzer = DataAnalyzer()

# Working with the set
analyzer.add_to_set([1, 2, 3, 4])
print("Set after addition:", analyzer.get_set())
analyzer.remove_from_set(2)
print("Set after removal:", analyzer.get_set())

# Working with the dictionary
analyzer.create_dictionary(["name", "age"], ["Alice", 30])
print("Dictionary:", analyzer.get_dictionary())
analyzer.update_dictionary("location", "Wonderland")
print("Updated Dictionary:", analyzer.get_dictionary())
print("Search for 'name':", analyzer.search_dictionary("name"))
analyzer.remove_from_dictionary("age")
print("Dictionary after removal:", analyzer.get_dictionary())
```
